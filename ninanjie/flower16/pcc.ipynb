```python
import matplotlib.pyplot as plt

# 模型及其测试集PCC值
models = ['Gradient Boosting', 'LightGBM', 'XGBoost', 'Random Forest']
pcc_values = [0.8286, 0.7898, 0.8594, 0.7893]

colors = ['#D3F1D1', '#ADD8E6', '#FBE6A2', '#CCFFFF']

plt.figure(figsize=(9, 6))
bars = plt.bar(models, pcc_values, color=colors, edgecolor='grey')

# 添加数值标签
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.01, f'{height:.2f}', 
             ha='center', va='bottom', fontsize=10)

plt.title('Comparison of Machine Learning Models (PCC)', fontsize=14, fontweight='bold')
plt.xlabel('Models', fontsize=12)
plt.ylabel('PCC (Pearson Correlation Coefficient)', fontsize=12)
plt.ylim(0, 0.9)
plt.tight_layout()
plt.show()
```
